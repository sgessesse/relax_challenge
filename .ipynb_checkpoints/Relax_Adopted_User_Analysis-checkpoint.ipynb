```python
# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import timedelta
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

# Load datasets
users_df = pd.read_csv('takehome_users.csv', encoding='ISO-8859-1')
engagement_df = pd.read_csv('takehome_user_engagement.csv')

# Convert the 'time_stamp' column to datetime
engagement_df['time_stamp'] = pd.to_datetime(engagement_df['time_stamp'], errors='coerce')

# Step 1: Identify adopted users
# Adopted user is defined as a user who logged in on three separate days within a seven-day period

# Step 2: Add a 'date' column to the engagement dataframe (ignoring time)
engagement_df['date'] = engagement_df['time_stamp'].dt.date

# Step 3: Group by user_id and date to get the number of unique login days
user_logins = engagement_df.groupby('user_id')['date'].nunique().reset_index()
user_logins.columns = ['user_id', 'login_days']

# Step 4: Identify users who logged in 3 times or more within any 7-day period
adopted_users = []
for user_id in engagement_df['user_id'].unique():
    # Filter data for this user and sort by timestamp
    user_data = engagement_df[engagement_df['user_id'] == user_id].sort_values('time_stamp')
    
    # Ensure the 'date' column is properly in datetime format
    user_data['date'] = pd.to_datetime(user_data['date'])
    
    # Calculate the difference in days between each login
    user_data['date_diff'] = user_data['date'].diff().dt.days
    
    # Initialize a counter for the number of logins in a 7-day window
    login_counter = 1
    for i in range(1, len(user_data)):
        if user_data.iloc[i]['date_diff'] <= 7:
            login_counter += 1
        else:
            login_counter = 1  # Reset counter if the gap is more than 7 days
        
        # If 3 or more logins occurred within a 7-day window, mark user as adopted
        if login_counter >= 3:
            adopted_users.append(user_id)
            break  # No need to check further for this user

# Step 5: Add an 'adopted_user' column to the users dataframe
users_df['adopted_user'] = users_df['object_id'].isin(adopted_users)

# Check if we have adopted users now
print(users_df['adopted_user'].value_counts())

# Plot the proportion of adopted vs non-adopted users
sns.countplot(x='adopted_user', data=users_df)
plt.title('Adopted vs Non-Adopted Users')
plt.show()

# Step 6: Feature selection
# Select relevant features: creation_source, opted_in_to_mailing_list, enabled_for_marketing_drip, org_id, invited_by_user_id
features = users_df[['creation_source', 'opted_in_to_mailing_list', 'enabled_for_marketing_drip', 'org_id', 'invited_by_user_id', 'adopted_user']]

# Convert categorical variables to dummy variables
features = pd.get_dummies(features, columns=['creation_source', 'org_id', 'invited_by_user_id'], drop_first=True)

# Separate the target variable and features
X = features.drop('adopted_user', axis=1)
y = features['adopted_user']

# Step 7: Train-test split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Step 8: Model training using RandomForestClassifier
model = RandomForestClassifier(random_state=42)
model.fit(X_train, y_train)

# Step 9: Model evaluation
y_pred = model.predict(X_test)
print('Accuracy:', accuracy_score(y_test, y_pred))
print('Confusion Matrix:\n', confusion_matrix(y_test, y_pred))
print('Classification Report:\n', classification_report(y_test, y_pred))

# Step 10: Feature importance
importances = model.feature_importances_
feature_names = X.columns
feature_importance_df = pd.DataFrame({'feature': feature_names, 'importance': importances}).sort_values(by='importance', ascending=False)

# Plot feature importance
plt.figure(figsize=(10, 8))
sns.barplot(x='importance', y='feature', data=feature_importance_df.head(20))
plt.title('Top 20 Feature Importance')
plt.xlabel('Importance')
plt.ylabel('Feature')
plt.tight_layout()
plt.show()
```


### Brief Write-up:

The analysis aimed to predict user adoption based on various features, using a Random Forest Classifier. Here's a summary of the approach and findings:

1. **Data Processing**: The "takehome_user_engagement.csv" and "takehome_users.csv" files were used. We defined an "adopted user" as one who logged in three times over any seven-day period. This was achieved by calculating a 7-day rolling count of logins for each user and labeling users who met the criteria.

2. **Exploratory Data Analysis (EDA)**: 
   - The dataset was not balanced in terms of adopted vs non-adopted users. There is a lot more non adopters compared to adopters
   - Key features for prediction included account creation source, opting into mailing lists, and marketing drip participation.

3. **Modeling and Evaluation**:
   - A Random Forest Classifier was chosen to predict user adoption.
   - The model achieved a reasonable accuracy on the test data, with key factors such as opting in to mailing list(`opted_in_to_mailing_list`) and marketing drip participation (`enabled_for_marketing_drip`) showing significant influence on adoption.

4. **Feature Importance**:
   - Opting in to mailing list(`opted_in_to_mailing_list`) was the most influential factor in predicting user adoption.
   - Users invited via organizations (`ORG_INVITE`) had higher chances of becoming adopted, suggesting that the context in which users are introduced to the platform plays a role.

### Recommendations for Further Research:

- Investigate **user behavior patterns** more deeply, such as the frequency and timing of logins beyond the seven-day period.
- Consider additional features such as **time since signup** and **interaction with other users** to better capture early user engagement trends.
